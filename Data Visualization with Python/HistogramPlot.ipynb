```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import os
from matplotlib import pyplot as plt

mpl.style.use('ggplot')
os.chdir('C:\\Users\\ppatrzalek\\Python Scripts\\')
```

```python
# Read dataset
df_canada = pd.read_csv('Datasets/df_canada.csv')
```

```python
# Set indexes
df_canada.set_index('Country', inplace=True)
```

```python
# Histogram (1)
df_canada['2013'].plot(kind='hist', figsize=(20, 10), alpha=0.8)

plt.title('Histogram of Immigration from 195 countries in 2013')
plt.ylabel('Number of Countries')
plt.xlabel('Number of Immigrants')

plt.show()
```

```python
# Histogram (2)
# Make better histogram using numpy
import numpy as np

count, bin_edges = np.histogram(df_canada['2013'], 
                                bins=10)

df_canada['2013'].plot(kind='hist', 
                       xticks=bin_edges, 
                       alpha=0.8, 
                       stacked=False,
                       figsize=(20, 10)
                      )
plt.title('Histogram of Immigration from 195 countries in 2013')
plt.ylabel('Number of Countries')
plt.xlabel('Number of Immigrants')

plt.show()
```
