```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import seaborn as sns
import os
from matplotlib import pyplot as plt

mpl.style.use('ggplot')
os.chdir('C:\\Users\\ppatrzalek\\Python Scripts\\')
```

```python
# Read dataset
csv_path = 'Datasets/df_total.csv'
df_total = pd.read_csv(csv_path)

csv_path2 = 'Datasets/df_canada.csv'
df_can = pd.read_csv(csv_path2)
```

```python
# Regression plot
ax = sns.regplot(x='year', y='total', data=df_total,
                color='green', marker='+')
```

```python
plt.figure(figsize=(15, 10))
ax = sns.regplot(x='year', y='total', data=df_total, color='green', marker='+', scatter_kws={'s':200})

ax.set(xlabel='Year', ylabel='Total Immigration')
ax.set_title('Total Immigration to Canada drom 1980 - 2013')
```

```python
plt.figure(figsize=(15, 10))

sns.set(font_scale=1.5)

ax = sns.regplot(x='year', y='total', data=df_total, color='green', marker='+', scatter_kws={'s': 200})
ax.set(xlabel='Year', ylabel='Total Immigration')
ax.set_title('Total Immigration to Canada from 1980 - 2013')
```

```python
plt.figure(figsize=(15, 10))

sns.set(font_scale=1.5)
#sns.set_style('ticks') # change background to white background
sns.set_style('whitegrid')

ax = sns.regplot(x='year', y='total', data=df_total, color='green', marker='+', scatter_kws={'s': 200})
ax.set(xlabel='Year', ylabel='Total Immigration')
ax.set_title('Total Immigration to Canada from 1980 - 2013')
```
