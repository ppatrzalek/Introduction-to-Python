```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import os
from matplotlib import pyplot as plt

mpl.style.use('ggplot')
os.chdir('C:\\Users\\ppatrzalek\\Python Scripts\\')
```

```python
# Read dataset
csv_path = 'Datasets/df_canada.csv'
df_canada = pd.read_csv(csv_path)
```

```python
# Create new dataframe
years = list(map(str, range(1980, 2013)))

df_total = df_canada.loc[:, years].transpose()
df_total['total'] = df_total.sum(numeric_only=True, axis=1)
df_total = df_total.loc[:, ['total']]
df_total.reset_index(inplace=True)
df_total.rename(columns={'index':'year'}, inplace=True)
df_total = df_total.astype({'year':'int64'})
df_total.head()
```

```python
# Scatter plot
df_total.plot(kind='scatter',
             x='year',
             y='total',
             figsize=(10,8))

plt.title('Total Immigrant population to Canada from 1980 - 2013')
plt.xlabel('Year')
plt.ylabel('Number of Immigrant')
plt.show()
```
