```python
import numpy as np
import pandas as pd
import os 
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline

os.chdir('C:\\Users\\ppatrzalek\\Python Scripts\\')
mpl.style.use('ggplot') # optional: for ggplot-like style
```

```python
# Read dataset
csv_path = 'Datasets/df_canada.csv'
df_canada = pd.read_csv(csv_path)
```

```python
# Dataset information
index = df_canada.index
columns = df_canada.columns
values = df_canada.values
```

```python
# Set index
df_canada.set_index("Country", inplace=True)
df_canada.head()
```

```python
# Generating Area Plots
years = list(map(str, range(1980, 2014)))

df_top5 = df_canada.head()
df_top5 = df_top5[years].transpose()
df_top5.rename(columns = {"United Kingdom of Great Britain and Northern Ireland": "United Kingdom"}, inplace=True)
df_top5.head()
```

```python
df_top5.plot(kind='area',
             stacked=False,
             figsize=(20,10), 
             alpha=0.45)

plt.title('Imigration Trend of Top 5 Countries')
plt.ylabel('Number of imigrants')
plt.xlabel('Years')

plt.show()
```
