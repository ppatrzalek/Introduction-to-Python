```python
import pandas as pd
import numpy as np
import matplotlib as mpl
from matplotlib import pyplot as plt
import os

os.chdir('C:\\Users\\ppatrzalek\\Python Scripts')
mpl.style.use('ggplot')
```

```python
# Read dataset
csv_path = 'Datasets/df_canada.csv'
df_canada = pd.read_csv(csv_path)

# Set Index
df_canada.set_index('Country', inplace=True)
```

```python
df_continents = df_canada.groupby('Continent', axis=0).sum()
df_continents.head()
```

```python
# Pie Chart
df_continents['Total'].plot(kind='pie', figsize=(8, 12))

plt.title('Immigration to Canada by Continent (1980-2013)')

plt.show()
```
